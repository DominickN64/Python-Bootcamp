```python
i = 0
TwoStarBooks = []
while i < 51:
    i = i+1
    bookstorepage = requests.get(url.format(i))
    bookstorepageSoup = bs4.BeautifulSoup(bookstorepage.text,"lxml")
    books = bookstorepageSoup.select(".product_pod")
    for book in books:
        if book.select(".star-rating.Two") != []:
            bookTitle = book.select("h3")[0].getText()
            TwoStarBooks.append(bookTitle)
            
            
            
            
```

```python
i = 0
ThreeStarBooks = []
while i < 51:
    i = i+1
    bookstorepage = requests.get(url.format(i))
    bookstorepageSoup = bs4.BeautifulSoup(bookstorepage.text,"lxml")
    books = bookstorepageSoup.select(".product_pod")
    for book in books:
        if book.select(".star-rating.Three") != []:
            bookUrl = book.select("h3 a")[0]["href"]
            fullUrl = "https://books.toscrape.com/catalogue/"+ bookUrl
            BookPage = requests.get(fullUrl)
            BookPageSoup = bs4.BeautifulSoup(BookPage.text,"lxml")
            bookTitle = BookPageSoup.select("h1")[0].getText()
            ThreeStarBooks.append(bookTitle)
            
```
